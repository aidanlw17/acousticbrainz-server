```python
import os

import gzip
import json

from collections import defaultdict
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
def scatter(metric, vectors, model, ids):
    data = {"x": vectors[:, 0], "y": vectors[:, 1], "model": model, "id": ids}
    df = pd.DataFrame(data=data)
    fig = px.scatter(df, hover_data=['model', 'id'], x="x", y="y", color='model')
    fig.show()
```

```python
# Import data
archive_path = "/Users/noods/Code/metabrainz/acousticbrainz/acousticbrainz-server/plots/similarity"
metrics = ["mfccs", "gfccs", "bpm", "instruments", "tzanetakis", "key", "mfccsw", "gfccsw", "dortmund", "rosamerica", "onsetrate", "moods"]
metric_dict = defaultdict(list)
for metric in metrics:
    name = metric + ".npy.gz"
    file = os.path.join(archive_path, name)
    metric_f = gzip.GzipFile(file, 'r')
    metric_dict[metric] = np.load(metric_f)

models = ["gender", "ismir04_rhythm", "timbre", "voice_instrumental", "danceability", "mood_acoustic", "mood_aggressive", "mood_electronic", "mood_happy", "mood_party", "mood_relaxed", "mood_sad", "moods_mirex", "genre_dortmund", "genre_rosamerica", "genre_tzanetakis", "genre_electronic"]
model_dict = defaultdict(list)
for model in models:
    name = model + "_model.npy.gz"
    file = os.path.join(archive_path, name)
    model_f = gzip.GzipFile(file, 'r')
    model_dict[model] = np.load(model_f)
    
name = "data_ids.npy.gz" 
file = os.path.join(archive_path, name)
f = gzip.GzipFile(file, 'r')
ids = np.load(f)
```

```python
scatter("gfccs", metric_dict["mfccs"], model_dict["genre_electronic"], ids)
```
